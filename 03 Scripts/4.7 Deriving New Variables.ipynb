### 02. Exercise

```python
# Create a subset
df = df_ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Define a function
def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] >15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
# If statement with loc
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

```python
# If statement with for loops
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
df_ords_prods_merged['busiest_day'] = result
```

### 03. Task

```python
# 02. Busiest and least busiest 2 days
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest days")
    elif value == 1:
        result.append("Busiest days")    
    elif value == 3:
        result.append("Slowest days")
    elif value == 4:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merged['busiest_days'] = result
```

```python
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

I noticed the busiest and slowest day counts nearly doubled, which makes sense since we are combining days

```python
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# 04. Identify busiest hours due to app freeze
result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
    if value == 10:
        result.append("Most orders")
    elif value == 11:
        result.append("Most orders")    
    elif value == 4:
        result.append("Fewest orders")
    elif value == 3:
        result.append("Fewest orders")
    else:
        result.append("Average orders")
```

```python
result
```

```python
df_ords_prods_merged['busiest_period_of_day'] = result
```

```python
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

the busiest hours of the day were based on top 2 hours(10 and 11 AM), and bottom 2(3 and 4 AM) for fewest orders. All other times were grouped into average orders
