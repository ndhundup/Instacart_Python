### 01. Importing

```python
# import libraries
import pandas as pd
import numpy as np
import os
```

```python
# import data
path = r'/Users/ngawangdhundup/Desktop/career foundry bootcamp/DM Achievement 4'
df_cust = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'customers.csv'), index_col = False)
df_orders_products_grouped = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_grouped.pkl'))
```

### Data wrangling

```python
df_cust.head()
```

```python
# Renameing column names
df_cust.rename(columns = {'STATE' : 'state'}, inplace = True )
df_cust.rename(columns = {'n_dependants' : 'number_of_dependants'}, inplace = True )
df_cust.rename(columns = {'fam_status' : 'family_status'}, inplace = True )
df_cust.rename(columns = {'First Name' : 'first_name'}, inplace = True )
df_cust.rename(columns = {'Surnam' : 'last_name'}, inplace = True )
df_cust.rename(columns = {'Gender' : 'gender'}, inplace = True )
df_cust.rename(columns = {'Age' : 'age'}, inplace = True )
```

```python
df_cust.shape
```

```python
df_cust.describe()
```

```python
df_cust.info()
```

```python
# Change user id into int
df_cust['user_id'] = df_cust['user_id'].astype('int64')
```

### Data Consistency

```python
# Missing values
df_cust.isnull().sum()
```

```python
# Check for mixed types

for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
df_cust['first_name'] = df_cust['first_name'].astype('str')
```

```python
df_cust.isnull().sum()
```

```python
df_cust.shape
```

```python
# Deleting nan from first name column that got changed due to mixed types
df_cust['first_name'].describe()
```

```python
df_cust[df_cust['first_name']=='nan']
```

```python
# Converting nan string back to null value, then dropping
df_cust_clean = df_cust.replace('nan', np.nan)
```

```python
df_cust_clean.dropna()
```

```python
# Finding duplicates
df_dups = df_cust_clean[df_cust_clean.duplicated()]
```
