#### Sampling 

```python
# Create a subset for line chart
df = ords_cust_merge[:2000000]
```

```python
np.random.seed(4)
```

```python
# Create a list holding True/False values to the test np.random.rant() <= 0.7
dev = np.random.rand(len(ords_cust_merge)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big
big = ords_cust_merge[dev]
```

```python
# Store 30% of the sample in the dataframe small
small = ords_cust_merge[~dev]
```

```python
len(ords_cust_merge)
```

```python
len(big) + len(small)
```

```python
df_2 = small[['orders_day_of_week','prices']]
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Exporting line chart
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_orders_dow.png'))
```

### 03. Task

```python
ords_cust_merge.head(5)
```

```python
# 03. Descriptive findings about sales via histogram
hist_sales = ords_cust_merge['order_hour_of_day'].plot.hist(bins = 75)
```

Majority of sales occur from 9 am to 4 pm. The outlier of sales are late at night/ early morning which makes sense since people are normally asleep during those hours. 

```python
# 04. Orders from loyal customers via bar chart
bar_loyal_cust = ords_cust_merge['loyalty_flag'].value_counts().plot.bar()
```

There are more orders from regular customers compared to loyal customers

```python
# 05. Difference in expenditure depending on hour of the day
df_3 = small[['order_hour_of_day','prices']]
```

```python
line_prices_by_hour = sns.lineplot(data = df_3, x = 'order_hour_of_day',y = 'prices')
```

There is a wide margin of error for the early hours of the day. While the middle of the day the pricing seems more consistent. For the sample the price range is limited in scope.

```python
# 07. Connection between age and income
df_4 = small[['age','income']]
```

```python
line_age_and_income = sns.lineplot(data = df_4, x = 'age',y = 'income')
```

The income drastically increases once the age crosses 40, which makes sense since you would make more money the longer you would be working
